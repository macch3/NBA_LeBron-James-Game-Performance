## Data Cleaning

```python
# Handling Missing Values
# Filling missing values in 'threep' (3-point percentage) and 'ftp' (free throw percentage)
lebron_data['threep'] = lebron_data['threep'].fillna(lebron_data['threep'].mean())
lebron_data['ftp'] = lebron_data['ftp'].fillna(lebron_data['ftp'].mean())

# Check for Duplicates
# Assuming that each row should be unique, we remove any duplicates
lebron_data = lebron_data.drop_duplicates()

# Data cleaning completed
print(lebron_data)
```

## EDA

```python
# Descriptive Statistics
descriptive_stats = lebron_data.describe()

# Displaying the descriptive statistics
descriptive_stats
```

```python
# Setting up the visualisation settings
sns.set(style="whitegrid")

# Distribution Analysis for key metrics
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

# Points distribution
sns.histplot(lebron_data['pts'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Points')

# Assists distribution
sns.histplot(lebron_data['ast'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Assists')

# Rebounds distribution
sns.histplot(lebron_data['trb'], kde=True, ax=axes[0, 2])
axes[0, 2].set_title('Distribution of Rebounds')

# Field Goal Percentage distribution
sns.histplot(lebron_data['fgp'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Field Goal Percentage')

# Three-Point Percentage distribution
sns.histplot(lebron_data['threep'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Three-Point Percentage')

# Free Throw Percentage distribution
sns.histplot(lebron_data['ftp'], kde=True, ax=axes[1, 2])
axes[1, 2].set_title('Distribution of Free Throw Percentage')

plt.tight_layout()
plt.show()

```
